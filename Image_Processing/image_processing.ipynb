```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
import cv2
img = cv2.imread('cute_cat.jpg')
img
```

```python
img.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

### CV2 considers image as BGR by default

```python
## Convert BGR to RGB
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img
```

```python
plt.imshow(img)
```

### Split the RGB component

```python
r,g,b = cv2.split(img)
```

```python
r
```

```python
r.shape
```

```python
g
```

```python
g.shape
```

```python
b
```

```python
b.shape
```

```python
plt.imshow(r,cmap='Reds')
```

```python
plt.imshow(img,cmap='Greens')
```

```python
plt.imshow(b,cmap='Blues')
```

### Gray scale 

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray,cmap='gray')
```
