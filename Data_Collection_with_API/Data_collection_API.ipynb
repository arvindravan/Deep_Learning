## Data collection using API

```python
import requests

url = "https://amazon-product-data6.p.rapidapi.com/product-by-text"

querystring = {"keyword":"iphone","page":"1","country":"US"}

headers = {
	"X-RapidAPI-Key": "985815fcb1msh694a01cb9f59ff8p1c125fjsn222cc29753f8",
	"X-RapidAPI-Host": "amazon-product-data6.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
data = response.json()
data
```

```python
len(data)
```

```python
for i,j in data.items():
    print(i,j)
```

```python
sel_data= data['data']
```

```python
sel_data[0]['asin']
```

```python
asin = []
title = []
price = []

for i,j in enumerate(sel_data):
    #print(i,j)
    asin.append(sel_data[i]['asin'])
    title.append(sel_data[1]['title'])
    price.append(sel_data[4]['price'])

    
```

```python
asin
```

```python
title
```

```python
price
```

### Convert that in datframe

```python
dct = {'asin':asin,
       'title':title,
       'price':price
       }
dct
```

```python
import pandas as pd
df = pd.DataFrame(dct)
df
```
